```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Veri kümesini yükleme
file_path = r'C:\Users\realb\Desktop\diyabet olasılığı/health care diabetes.csv'
data = pd.read_csv(file_path)

# İlk birkaç satırı görüntüleme
print(data.head())

# Temel istatistikleri hesaplama
print(data.describe())

# Boş değerleri ele alma
# 0 değerleri olan sütunları kontrol etme
columns_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# 0 değerlerini NaN ile değiştirme
data[columns_with_zeros] = data[columns_with_zeros].replace(0, np.nan)

# Her sütundaki boş değer sayısını görüntüleme
print(data.isnull().sum())

# Boş değerleri sütunların medyanı ile doldurma
data = data.fillna(data.median())

# Boş değerlerin yeniden kontrol edilmesi
print(data.isnull().sum())

# Temel istatistikleri yeniden hesaplama
print(data.describe())

# Veriyi görselleştirme
# Yaşa göre diyabet dağılımı
plt.figure(figsize=(10, 6))
sns.histplot(data, x='Age', hue='Outcome', multiple='stack')
plt.title('Age Distribution by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Glikoz ve BMI arasındaki ilişki
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=data)
plt.title('Glucose vs BMI by Diabetes Outcome')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()

# Özellikler arasındaki korelasyon matrisi
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```
