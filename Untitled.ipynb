```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Veri kmesini ykleme
file_path = r'C:\Users\realb\Desktop\diyabet olasl/health care diabetes.csv'
data = pd.read_csv(file_path)

# lk birka satr grntleme
print(data.head())

# Temel istatistikleri hesaplama
print(data.describe())

# Bo deerleri ele alma
# 0 deerleri olan stunlar kontrol etme
columns_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# 0 deerlerini NaN ile deitirme
data[columns_with_zeros] = data[columns_with_zeros].replace(0, np.nan)

# Her stundaki bo deer saysn grntleme
print(data.isnull().sum())

# Bo deerleri stunlarn medyan ile doldurma
data = data.fillna(data.median())

# Bo deerlerin yeniden kontrol edilmesi
print(data.isnull().sum())

# Temel istatistikleri yeniden hesaplama
print(data.describe())

# Veriyi grselletirme
# Yaa gre diyabet dalm
plt.figure(figsize=(10, 6))
sns.histplot(data, x='Age', hue='Outcome', multiple='stack')
plt.title('Age Distribution by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Glikoz ve BMI arasndaki iliki
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=data)
plt.title('Glucose vs BMI by Diabetes Outcome')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()

# zellikler arasndaki korelasyon matrisi
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```
